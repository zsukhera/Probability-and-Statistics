Finding the number of missing values and replacing them

```python
# Check for missing values
missing_values = df.isnull().sum()

print(f"\nTotal missing values: {df.isnull().sum().sum()}")
print(f"Percentage of missing data: {(df.isnull().sum().sum() / (df.shape[0] * df.shape[1])) * 100:.2f}%")

# Identify columns with missing values
cols_with_missing = df.columns[df.isnull().any()].tolist()
print(f"\nColumns with missing values: {cols_with_missing}")

# Replace missing values with mean for numeric columns
for col in cols_with_missing:
    if df[col].dtype in ['float64', 'int64']:
        mean_value = df[col].mean()
        df[col].fillna(mean_value, inplace=True)
        print(f"Filled {col} with mean: {mean_value:.2f}")


# For categorical columns (Ethnicity), use mode
if 'Ethnicity' in cols_with_missing:
    mode_value = df['Ethnicity'].mode()[0]
    df['Ethnicity'].fillna(mode_value, inplace=True)
    print(f"Filled Ethnicity with mode: {mode_value}")
```

Outlier Detection and Removal

```python
def detect_outliers_iqr(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return (series < lower) | (series > upper)

iqr_outliers = {c: detect_outliers_iqr(df[c]).sum() for c in ['Age', 'GPA', 'StudyTimeWeekly', 'Absences'] }
iqr_outliers


#Remove rows that are outliers in any of a chosen list
cols_for_outlier = ['StudyTimeWeekly','Absences','GPA']
mask_outlier = np.any([detect_outliers_iqr(df[c]) for c in cols_for_outlier], axis=0)
df_iqr_removed = df.loc[~mask_outlier].copy()
print("Removed rows:", mask_outlier.sum())
```

Pie Charts of Categorical Variables

```python
# helper to map numeric categorical codes to readable labels
par_edu_map = {0:'None',1:'High School',2:'Some College',3:"Bachelor's",4:'Higher'}
parent_support_map = {0:'None',1:'Low',2:'Moderate',3:'High',4:'Very High'}
gender_map = {0:'Male',1:'Female'}
tf_map = {0:'No',1:'Yes'}

# create label columns for plotting
df['ParentalEducation_label'] = df['ParentalEducation'].round().astype(int).map(par_edu_map)
df['ParentalSupport_label'] = df['ParentalSupport'].round().astype(int).map(parent_support_map)
df['Gender_label'] = df['Gender'].round().astype(int).map(gender_map)
df['Tutoring_label'] = df['Tutoring'].round().astype(int).map(tf_map)
df['Extracurricular_label'] = df['Extracurricular'].round().astype(int).map(tf_map)
df['Sports_label'] = df['Sports'].round().astype(int).map(tf_map)
df['Music_label'] = df['Music'].round().astype(int).map(tf_map)
df['Volunteering_label'] = df['Volunteering'].round().astype(int).map(tf_map)

# pie plotting function
def plot_pie(column_label, title):
    counts = df[column_label].value_counts()
    plt.figure(figsize=(6,6))
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)
    plt.title(title)
    plt.axis('equal')
    plt.show()

# create pies
plot_pie('Ethnicity_label', 'Ethnicity Distribution')
```

Pareto Chart

```python
eth_counts = df['Ethnicity_label'].value_counts().sort_values(ascending=False)
cumperc = eth_counts.cumsum()/eth_counts.sum()*100

fig, ax1 = plt.subplots(figsize=(8,5))
ax1.bar(eth_counts.index, eth_counts.values)
ax1.set_ylabel('Count')
ax2 = ax1.twinx()
ax2.plot(eth_counts.index, cumperc.values, color='C1', marker='D', ms=7)
ax2.set_ylabel('Cumulative %')
ax2.axhline(80, color='grey', linestyle='--', linewidth=0.8)
plt.title('Pareto Chart: Ethnicity')
for tick in ax2.get_yticks():
    pass
plt.show()
```
